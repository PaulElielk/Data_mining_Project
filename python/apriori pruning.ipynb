```python
import pandas as pd

# Load transactions
df = pd.read_csv("coin_transactions.csv")

# Build basket (transaction_id x car_id)
basket = df.groupby(["transaction_id"])["car_id"].apply(list)
basket = basket.tolist()
basket
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(basket).transform(basket)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df_encoded, min_support=0.0005, use_colnames=True)

```

```python
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, 
                          metric="confidence", 
                          min_threshold=0.2)
```

```python
rules
```

```python
rules["left_car_id"] = rules["antecedents"].apply(lambda x: list(x)[0])
rules["right_car_id"] = rules["consequents"].apply(lambda x: list(x)[0])
```

```python
rules = rules[
    (rules["antecedents"].apply(len) == 1) &
    (rules["consequents"].apply(len) == 1)
]
```

```python
final_rules = rules[[
    "left_car_id",
    "right_car_id",
    "support",
    "confidence",
    "lift"
]]
```

```python
final_rules.to_csv("coin_rules_for_site.csv", index=False)
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Load NEW Jumia transactions
df = pd.read_csv("jumia_transactions.csv")

# Convert to basket list
baskets = df.groupby("transaction_id")["product_id"].apply(list).tolist()

# One-hot encode
te = TransactionEncoder()
te_ary = te.fit(baskets).transform(baskets)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# Apriori: start with reasonable support
frequent_itemsets = apriori(
    df_encoded,
    min_support=0.02,    # 2% of 1000 = appears in at least 20 transactions
    use_colnames=True
)

print("Frequent itemsets:", len(frequent_itemsets))

if len(frequent_itemsets) == 0:
    print("⚠️ No itemsets, lower min_support to 0.01 or 0.005 before panicking.")
```

```python
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.01   # 10% confidence
)

print("Total rules before filtering:", len(rules))

# Keep only 1 → 1 rules
rules = rules[
    (rules["antecedents"].apply(len) == 1) &
    (rules["consequents"].apply(len) == 1)
]

print("Rules after filtering:", len(rules))

rules["left_product_id"] = rules["antecedents"].apply(lambda x: list(x)[0])
rules["right_product_id"] = rules["consequents"].apply(lambda x: list(x)[0])

final_rules = rules[[
    "left_product_id",
    "right_product_id",
    "support",
    "confidence",
    "lift"
]]

final_rules.to_csv("jumia_rules_for_site_NEW.csv", index=False)
print("Saved jumia_rules_for_site_NEW.csv with", len(final_rules), "rules")
```

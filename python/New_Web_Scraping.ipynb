### Web Scrap for CoinAfrique ###

```python

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time

#  Configuration of navigator Chrome 
options = Options()
options.add_argument("--headless")           # # No browser display
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

service = Service("C:/Users/PC/Documents/Datascience/Data Mining/Final project/chromedriver-win64/chromedriver.exe")
driver = webdriver.Chrome(service=service, options=options)

# --- Parameters ---
max_pages = 20   
base_url = "https://ci.coinafrique.com/categorie/vehicules?page="

#  Stock datas 
sellers, countries, descriptions, prices, images = [], [], [], [], []

#  loop sur plusieurs pages
for page in range(1, max_pages + 1):
    url = base_url + str(page)
    print(f"ðŸ“„ Scraping page {page} â†’ {url}")
    driver.get(url)
    time.sleep(3)  
    
    soup = BeautifulSoup(driver.page_source, "html.parser")
    cards = soup.find_all("div", class_="card ad__card round small hoverable undefined")
    
    print(f"  â†’ Found {len(cards)} ads")

    for card in cards:
        # Seller
        seller = card.find("p", class_="title")
        sellers.append(seller.text.strip() if seller else None)

        # Country / Location
        country_tag = card.find("p", class_="ad__card-location")
        country = country_tag.find("span").text.strip() if country_tag and country_tag.find("span") else None
        countries.append(country)

        # Description / Title
        desc_tag = card.find("p", class_="ad__card-description")
        desc_text = desc_tag.find("a").text.strip() if desc_tag and desc_tag.find("a") else (desc_tag.text.strip() if desc_tag else None)
        descriptions.append(desc_text)

        # Price
        price_tag = card.find("p", class_="ad__card-price")
        prices.append(price_tag.text.strip() if price_tag else None)

        # Image
        img_tag = card.find("img")
        images.append(img_tag["src"] if img_tag and "src" in img_tag.attrs else None)

#  close  navigator 
driver.quit()

# --- Creation du DataFrame ---
df_coin = pd.DataFrame({
    "Seller": sellers,
    "Country": countries,
    "Description": descriptions,
    "Price": prices,
    "Image": images
})

print(f"\n Total annonces rÃ©cupÃ©rÃ©es : {len(df_coin)}")
df_coin.head()
```

```python
import pandas as pd


df= pd.read_csv('Jumia Newdataset.csv')

#  only  needed columns
df = df[["Brand_name", "Price", "Image"]]

#   clean price 
def clean_price(p):
    p = str(p).strip()
    p = p.replace(" ", "")  # remove all spaces
    
    if "-" in p:  
        # take minimum 
        return float(p.split("-")[0])
    
    return float(p)

#  Apply cleaning
df["Price"] = df["Price"].apply(clean_price)

#  Reorder columns just in case 
df = df[["Brand_name", "Price", "Image"]]

#  Save final clean CSV 
df.to_csv("Jumia_FINAL.csv", index=False)

print("âœ… Jumia_FINAL.csv cleaned and saved!")
```
